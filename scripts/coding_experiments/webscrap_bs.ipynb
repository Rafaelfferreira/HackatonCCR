```python
table = {
    'Posto': [],
    'Endereo': [],
    'Fone': [],
    'Obs': [],
    'Items': []
}
```

```python
def is_another_item(string):
    return not string.isupper()

def grab_all_items(i, cels):
    
    items_str = ""
    while is_another_item(cels[i]['title']):
        items_str += cels[i]['title'] + "&"
        i+=1
        
        if(i >= len(cels)):
            break
    
    return items_str[:-1], i
    
```

```python
total_cels = len(columns)

i = 0
while i < total_cels:
    
    table['Posto'].append(columns[i]['title'])
    table['Endereo'].append(columns[i+1]['title'])
    table['Fone'].append(columns[i+2]['title'])
    table['Obs'].append(columns[i+3]['title'])
    
    items, i = grab_all_items(i+4, columns)
    table['Items'].append(items)
    
    
```
